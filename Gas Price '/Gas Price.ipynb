##  1st Insights:Cost of Gas in Different Countries with respect to it's year?

```python
gas.groupby(["Australia","Canada","France","Germany","Italy","Japan","Mexico","South Korea","UK","USA"])['Year'].max()
```

```python
plt.figure(figsize=(10,10),dpi=100)
plt.title("Gas Cost of Different Countries")
plt.xlabel("Year")
plt.ylabel("Cost")

plt.plot(gas['Year'],gas['Australia'],label="Australia",marker='*')
plt.plot(gas['Year'],gas['Canada'],label="Canada",marker='o')
plt.plot(gas['Year'],gas['France'],label="France",marker='o')
plt.plot(gas['Year'],gas['Germany'],label="Germany",marker='o')
plt.plot(gas['Year'],gas['Italy'],label="Italy",marker='o')
plt.plot(gas['Year'],gas['Japan'],label="Japan",marker='o')
plt.plot(gas['Year'],gas['Mexico'],label="Mexico",marker='o')
plt.plot(gas['Year'],gas['South Korea'],label="South Korea",marker='o')
plt.plot(gas['Year'],gas['UK'],label="UK",marker='o')
plt.plot(gas['Year'],gas['USA'],label="USA",marker='o')
plt.xticks(gas['Year'][::3])
plt.legend()
plt.show()
```
